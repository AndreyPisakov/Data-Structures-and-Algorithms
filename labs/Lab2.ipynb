```python
#Бинарный поиск
import random
def binary_search_recursive(array, element: int, start: int, end: int, add: bool = False) -> int:
    if start > end:
        if add:
            if element > array[start]:
                start += 1
            return start
        else:
            return -1
    mid = (start + end) // 2
    if element == array[mid]:
        return mid
    if element < array[mid]:
        return binary_search_recursive(array, element, start, mid-1, add)
    else:
        return binary_search_recursive(array, element, mid+1, end, add)


def add(array, elem):
    i = binary_search_recursive(array, elem, 0, len(array) - 1, True)
    array.insert(i, elem)


def delete(array, elem):
    i = binary_search_recursive(array, elem, 0, len(array) - 1)
    if i != -1:
        array.pop(i)
        
        
n = 5
mas = [random.randint(-1000, 1000) for j in range(n)]
mas.sort()
print(mas)
elem1 = int(input("Введите элемент для поиска"))
elem2 = int(input("Введите элемент для вставки"))
elem3 = int(input("Введите элемент для удаления"))
print("index", binary_search_recursive(mas, elem1, 0, len(mas) - 1))
add(mas, elem2)
delete(mas, elem3)
print(mas)
```

```python
#Метод Фибоначчи
class Fib:
    def __init__(self):
        self.i = 0
        self.p = 0
        self.q = 0
        self.stop = False

    def get_fib_number(self, k):
        first_num = 0
        second_num = 1
        for i in range(k):
            t = second_num
            second_num += first_num
            first_num = t
        return first_num

    def start_init(self, array):
        self.stop = False
        k = 0
        while self.get_fib_number(k + 1) < len(array):
            k += 1
        m = self.get_fib_number(k + 1) - len(array) - 1
        self.i = self.get_fib_number(k) - m
        self.p = self.get_fib_number(k - 1)
        self.q = self.get_fib_number(k - 2)

    def up_index(self):
        if self.p == 1:
            self.stop = True
        self.i += self.q
        self.p -= self.q
        self.q -= self.p

    def down_index(self):
        if self.q == 0:
            self.stop = True
        self.i -= self.q
        t = self.q
        self.q = self.p - self.q
        self.p = t

    def search(self, array, elem, add: bool = False):
        self.start_init(array)
        while not self.stop:
            if self.i < 0:
                self.up_index()
            elif self.i >= len(array):
                self.down_index()
            elif add and self.i != 0 and array[self.i] > elem >= array[self.i - 1]:
                return self.i
            elif add and self.i != len(array) - 1 and array[self.i] < elem <= array[self.i + 1]:
                return self.i + 1
            elif add and self.i == 0 and array[self.i] > elem:
                return self.i
            elif add and self.i == len(array) - 1 and array[self.i] < elem:
                return self.i
            elif not add and array[self.i] == elem:
                return self.i
            elif elem < array[self.i]:
                self.down_index()
            elif elem > array[self.i]:
                self.up_index()
        return -1

    def add(self, array, elem):
        i = self.search(array, elem, True)
        array.insert(i, elem)

    def delete(self, array, elem):
        i = self.search(array, elem)
        if i != -1:
            array.pop(i)


fs = Fib()
n = 10
array = [random.randint(-1000, 1000) for j in range(n)]
array.sort()
print(array)
elem1 = int(input("Введите элемент для поиска"))
elem2 = int(input("Введите элемент для вставки"))
elem3 = int(input("Введите элемент для удаления"))
print("index", fs.search(array, elem1))
fs.add(array, elem2)
fs.delete(array, elem3)
print(array)
```

Написать программу,  которая находит хотя бы один способ решения задач.


```python
global queen
global number
queen = [None] * 8
number = 0


def print_table():
    global number
    number += 1
    print('\n%d решение проблемы восьми ферзей \t' % number)
    for x in range(8):
        for y in range(8):
            if x == queen[y]:
                print('<q>', end=' ')
            else:
                print('<->', end=' ')
        print('\t')
    #input('\n.. нажмите любую клавишу, чтобы продолжить ..')


def attack(row, col):
    global queen
    for i in range(col):
        offset_col = abs(i - col)
        offset_row = abs(queen[i] - row)
        if queen[i] == row or offset_row == offset_col:
            return True
    return False


def decide_position(value):
    global queen
    for i in range(8):
        if not attack(i, value):
            queen[value] = i
            if value == 7:
                print_table()
            else:
                decide_position(value + 1)


decide_position(0)
```
